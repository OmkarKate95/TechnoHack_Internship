## Task 3 :- Visualization using Histogram

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
```

### Importing Data

```python
iris = pd.read_csv("Iris.csv")
iris
```

```python
iris.info()
```

```python
iris.isnull().sum()
```

## Data Visualization

```python
sns.histplot(data=iris, x="SepalWidthCm", hue="Species", kde=True)
plt.title("Histogram of Sepal Width for each Species")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Density")
```

```python
iris.hist(column=["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"])
plt.show()
```

```python
plt.hist(iris[iris["Species"] == "Iris-setosa"]["PetalWidthCm"], label="Iris-setosa")
plt.hist(iris[iris["Species"] == "Iris-versicolor"]["PetalWidthCm"], label="Iris-versicolor")
plt.hist(iris[iris["Species"] == "Iris-virginica"]["PetalWidthCm"], label="Iris-virginica")

plt.title("Histogram of Petal Width for each Species")
plt.xlabel("Petal Width (cm)")
plt.ylabel("Frequency")
plt.legend()
```

```python
plt.hist(iris[iris["Species"] == "Iris-setosa"]["SepalWidthCm"], label="Iris-setosa")
plt.hist(iris[iris["Species"] == "Iris-versicolor"]["SepalWidthCm"], label="Iris-versicolor")
plt.hist(iris[iris["Species"] == "Iris-virginica"]["SepalWidthCm"],  label="Iris-virginica")

plt.title("Histogram of Sepal Width grouped by Species")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
plt.legend()
```

```python
plt.hist2d(iris["SepalLengthCm"], iris["SepalWidthCm"],cmap="Reds")

plt.title("Histogram of Sepal Length and Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
```

```python
plt.hist(iris[iris["Species"] == "Iris-setosa"]["SepalLengthCm"], label="Iris-setosa")
plt.hist(iris[iris["Species"] == "Iris-versicolor"]["SepalLengthCm"], label="Iris-versicolor")
plt.hist(iris[iris["Species"] == "Iris-virginica"]["SepalLengthCm"], label="Iris-virginica")

plt.title("Histogram of Sepal Length for each Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.legend()
```
